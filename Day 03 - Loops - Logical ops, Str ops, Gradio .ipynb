## Logical Operations
- Boolean logic system - electronics, mathematics, number systems
- exists on two symbol number systems
- Binary Number System
  - 0 , 1
  - False, True
  - OFF, ON
  - 0 Volts, 5 Volts

- Logical operations came to existance for Bool values
- might work for other data types, but are most important with Booleans

- Operators:

OR

AND

NOT

```python
print("Truth Table for ---- OR operation")
print("False False ---> ", False or False)
print("False True ---> ", False or True)
print("True False ---> ", True or False)
print("True True ---> ", True or True)

print("--"*10)
print("Truth Table for ---- AND operation")
 
print("False False ---> ", False and False)
print("False True ---> ", False and True)
print("True False ---> ", True and False)
print("True True ---> ", True and True)
```

```python
# take an input from the user
# let's say we have to do a google search
# or we have to prompt an LLM - ChatGPT, or Gemini etc

prompt = input("Enter your query:")
print(prompt)

# string operations help us clean the users inputs
```

```python
print(prompt)

print(prompt.lstrip()) # removes the leading spaces - LEFT
print(prompt.rstrip()) # RIGHT - removes trailing spaces

print(prompt.strip()) # removes useless spaces on left or right

print(prompt.upper())
print(prompt.lower())

# search for something
prompt.find("awesome") # -1 if not found
prompt.find('who') # returns the index of the word

prompt.count("w")
prompt.count("who")


```
