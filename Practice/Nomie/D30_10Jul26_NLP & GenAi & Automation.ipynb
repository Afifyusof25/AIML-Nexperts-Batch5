### Exercise:

> Exercise:
1. Ask the user for prompt
2. create a function that caculates takes user input prompt
    - lowercase, tokenize
    - calcualte the initial number of tokens used
    - stop word removal
    - calculate new number of tokens
    - set_limit = 50
    - give user a warning once the limit is reached
    - implement it on the prompt
    - Say "here is the answer", or "say token limit reached, Recharge!!"
3. this function will take user's input as prompt

> 25 Minutes

```python
### Exercise: Solution

print("Please insert your prompt")
user_input = input("Ask me a question: ")
 
def prompt(user_input):
 
    # Original Prompt
    print("Initial prompt by user:")
    print("-"*30)
    print(f"{user_input}")
    print("="*30)
    
 
    # Lowercase
    newtext = user_input.lower()
 
    # Tokenize
    newtext_token = user_input.split()
 
    # Number of token
    newtext_number_of_token = len(newtext_token)
    
    print(f"Number of token {newtext_number_of_token}")
    print("="*30)
 
    # Stop word removal
    stopwords = ['the', 'a', 'an', 'in', 'on', 'at', 'and', 'or', 'but',
                 'i', 'you', 'he', 'she', 'is', 'am', 'are']
   
    clean_newtext = [word for word in newtext_token if word not in stopwords]
 
    # Calculate new tokens
    newtext_token_after_removal = len(clean_newtext)
    
    print(f"Number of token after removal of stop words: {newtext_token_after_removal}")
    print("="*30)
 
    # Set limit = 50
    set_limit = 50
 
    # warning
    if newtext_token_after_removal < set_limit:
        print("="*30)
        print("Looking for answers, wait for the response...")
        print("="*30)
        print(f"Here are the answer: ")
        print("="*30)
        print(f"{clean_newtext}")
        print("="*30)
 
    else:
        print("="*30)
        print("\nToken has reached limit. Suscribe our plan for further usage")
 
prompt(user_input)

```

```python
### STEMMING VS LEMMATIZATION

import nltk

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

sentence = 'Forests are breathing and evolving worlds that shelter countless species.'

stemmed_words = [stemmer.stem(word) for word in sentence.split()]

# stemmer.stem('eg. word')

print(" ".join(stemmed_words))

```

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

sentence = 'Forests are breathing and evolving worlds that shelter countless species.'
words = sentence.lower().split()

lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in words]

print(" ".join(lemmatized_words))



```
