```python
df2 = df.copy()
```

```python
df2['Price'].describe()
```

```python
Q1 = df2['Price'].quantile(0.25)
Q3 = df2['Price'].quantile(0.75)

print(f"Q1 : ", Q1)
print(f"Q3 : ", Q3)

IQR = Q3 - Q1
print(f'IQR : ', IQR)

# to detect outliers

Lower_Price = Q1 - 1.5*IQR
print("Lower Price : ", Lower_Price)

Upper_Price = Q3 + 1.5*IQR
print("Upper Price : ", Upper_Price)

outliers = df2[(df2['Price'] < Lower_Price) | (df2['Price'] > Upper_Price)]
print(outliers)


# to remove outliers

df2 = df2[(df2['Price'] >= Lower_Price) &
    (df2['Price'] <= Upper_Price)].reset_index(drop=True)



sns.boxplot(data = df2, x = 'Price')

sns.boxplot(df2)




```
