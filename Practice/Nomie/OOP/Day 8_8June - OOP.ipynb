```python
def summing(num1,num2,power):
    sum = (num1+num2) ** power
    return sum

# function cell
summing(1,-2,3) # order is important

# keyword arguments 
summing(power=3, num1=3,num2=1) # order is not important

summing(1,-3, 3)

# with default arguments
def summing(num1=1,num2=1,power=1):
    sum = (num1+num2) ** power
    return sum

summing(1,-4,2)
```

```python
## exercise 2 - OOP

class Dog:
    # class attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age
   
   # class method
    def sound(self):
        print("Wooofff!!!")
   
    def details(self):
        print(f"This is: {self.name}\n Age is: {self.age}")
 
class Cat:
    # class attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age
 
    # class method
    def sound(self):
        print("Meeowwww!!!")
   
    def details(self):
        print(f"This is: {self.name}\n Age is: {self.age}")
 
class Animal:
    # class attributes
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self.age = age
 
    # class method
    def details(self):
        print(f'This is {self.name}\n A {self.species}\n Age is: {self.age}')
 
# Objects
dog1 = Dog("Mochi", 3)
cat1 = Cat("Coco", 2)
animal1 = Animal("Mocha","Bird", 1)
 
# Outputs
dog1.sound()
dog1.details()

cat1.sound()
cat1.details()
 
animal1.details()
```

```python
# Parent class
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self._age = age 

    def introduce(self):
        print(f"Hi. I am {self.name}!!\nI am a {self.species}\nI am {self._age} years old!")

    # Getter Method
    def get_age(self):
        return self._age
    
    # Setter Method
    def set_age(self, new_age):
        self._age = new_age
        return f"The new age is {new_age}"
    
    def have_birthday(self):
        self._age +=1
        print(f"It's {self.name}'s birthday. Now {self._age} years old ")


    
```

```python
# CHILD of ANIMAL
class Dog(Animal):

    def __init__(self, name, age): # this is what we will pass ourselves
        # super() refers to the parent class --- could have used Animal as well
        # 
        self._age = age
        super().__init__(name, "Dog", age)

    def sound(self):
        print(f"{self.name} says: Woof!!")

    # different methods for different classes
    def detect_drugs(self):
        if self._age > 5:
            return f"I will sniff out any drugs"
        else:
            return f"{self.name} says, I am just {self.age} years old. You fr??"
    
```

```python
# CHILD of ANIMAL
class Cat(Animal):

    def __init__(self, name, age):
        super().__init__(name,"Cat", age)

    def sound(self):
        print(f"{self.name} says: Meowww!!")
```
