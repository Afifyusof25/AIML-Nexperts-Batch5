```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Load the dataset

data = {
    'CustomerID': [1, 2, 3, 4, 5, 6],
    'Age': [25, 26, 60, 62, 22, 65],
    'SpendingScore': [80, 85, 15, 10, 88, 20]
    }

df = pd.DataFrame(data)
print("--- Original Data ---")
print(df.head())
```

```python
# Feature selection - features that we want to use for clustering

X = df[['Age', 'SpendingScore']]
```

```python
# Create a Dendrogram to visualize the hierarchical clustering

# 1. Linkage
Z = linkage(X, method='ward')

# 2. Dendrogram
dendrogram(Z, labels=df['CustomerID'].values)

plt.axhline(y=15, color='r', linestyle='--')  # Example threshold line


#plt.figure(figsize=(10, 7))

#plt.title('Dendrogram')
#plt.xlabel('Data Points')
#plt.ylabel('Euclidean Distances')
plt.show()
```

```python
# Hierarchical Model

hc_model = AgglomerativeClustering(
    n_clusters=2,
    metric='euclidean',
    linkage='ward'
    )
```
