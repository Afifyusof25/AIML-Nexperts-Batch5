```python
import numpy as np

inputs =  np.array([1, 1, 1, 0, 0]) # depends on situation # can be anything
weights = np.array([0.7, 0.5, 0.6, 0.3, 0.4 ]) # we assign weights ourselves

# let's apply the frankrosenblat algo 
# Let's use a Perceptron to make our decision

# Step 1: Threshold
threshold = 1.5

# Step 2: Multiply inputs with weights
product = inputs * weights # broadcasting # [0.7 , 0, 0.6, 0, 0]

# Step 3: Weighted sum
weighted_sum = sum(product)

# Step 4: activate the perceptron
def step_activation(x):
    if x >= threshold:
        return "Yes, We will go to the Concert!"
    else:
        return "No. We ain't going."

# apply the activation
step_activation(weighted_sum)
```

```python
# another implementation
# combined in one function

def perceptron(artist, weather, friend, food, juice, threshold = 1.5):
    inputs = np.array([artist, weather, friend, food, juice])
    # the neuron already has the importances
    weights = np.array([0.7, 0.3, 0.8, 0.3, 0.4])

    # step2: multiply the inputs and weights
    product = inputs * weights

    # Step3: weighted sum
    weighted_sum = sum(product)

    # step4: activate the output
    if weighted_sum >= 1.5:
        return "Yes, we go!!!"
    else: 
        return "NOPE, we are not going!"


# whether is good
decision_sunny = perceptron(artist = 0, weather = 1, friend = 1, food = 1, juice = 1)
decision_sunny
```
