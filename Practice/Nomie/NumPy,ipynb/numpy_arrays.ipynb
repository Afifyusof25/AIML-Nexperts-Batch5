### Exercises - Scenario 1

The Scenario:

- The school's database has just exported a batch of raw exam scores. You need to use NumPy to generate some sample data, inspect its structure, and automatically grade the students based on a curve.

Your Tasks:

1. The Setup: Import NumPy. To ensure the whole class gets the exact same "random" numbers, set the random seed to 42.

2. Data Generation: Use random.randint to generate a 2D array of random scores between 0 and 100. The array should represent 10 students and their scores across 3 different subjects.

3. Array Inspection: Prove you built the array correctly by printing its ndim (number of dimensions), its shape, and its total size (total number of scores).

4. Student IDs: Use arange to create a 1D array of Student IDs starting at 100 and ending at 110.

5. Perform the aggregate operation of average and find the average score of students for each subject

6. The Audit (Boolean Masking): The principal wants to know every instance where a student scored below a 50. Use boolean masking to extract and print just those failing scores from the main array.

7. Automated Grading (where): Use np.where to create a brand new "Report Card" array. If a score in the original array is less than 50, it should say "Fail". If it is 50 or above, it should say "Pass".

```python
## 1. The Setup: Import NumPy. To ensure the whole class gets the exact same "random" numbers, set the random seed to 42.

import numpy as np
```

```python
# 1. Setup
np.random.seed(42)
```

```python
# 2. Data Generation: Use random.randint to generate a 2D array of random scores between 0 and 100. 
# The array should represent 10 students and their scores across 3 different subjects.

scores = np.random.randint(0, 101, (10, 3))

print("Scores Array:\n", scores)

```

```python
# 3. Array Inspection: Prove you built the array correctly 
# by printing its ndim (number of dimensions), its shape, and its total size (total number of scores).

print("Array Dimensions (ndim):", scores.ndim)
print("Array Shape:", scores.shape)
print("Total Size:", scores.size)
```

```python
# 4. Student IDs: Use arange to create a 1D array of Student IDs starting at 100 and ending at 110.

student_ids = np.arange(100, 111)
print("Student IDs:\n", student_ids)
```

```python
# 5. Perform the aggregate operation of average and find the average score of students for each subject

subject_avg = np.mean(scores, axis=0)
print("\nAverage Score per Subject:", subject_avg)
```

```python
# 6. The Audit (Boolean Masking): The principal wants to know every instance where a student scored below a 50. 
# Use boolean masking to extract and print just those failing scores from the main array.

fail_scores = scores[scores < 50]
print("All Failing Scores Below 50:", fail_scores)
```

```python
# 7. Automated Grading (where): Use np.where to create a brand new "Report Card" array. 
# If a score in the original array is less than 50, it should say "Fail". If it is 50 or above, it should say "Pass".

report_card = np.where(scores < 50, "Fail", "Pass")

print("Scores Array:\n", scores)
print("\nReport Card:\n", report_card)

print("\n--- Detailed Report ---")

zip_list = list(zip(scores, report_card))

for Scores_Array, Report_Card in zip(scores, report_card):
    print(f"{Scores_Array} : {Report_Card}")

```

### Exercises - Scenario 2

Scenario 2:

The semester is over. The principal wants a final report that includes Student IDs, their final scores, and a text label indicating whether they Passed or Failed. You need to combine all this data into one final grid and export it so it can be sent to the administration.

Your Tasks:

1. The Numerical Data: Access the 1D marks arrays of 5 Student IDs (101 to 105), store separately in variables . 

2. The Text Data: Use `np.where` to create a status array. If the score is 50 or above, the status is "Pass". If it's below 50, it is "Fail".

3. The Merge (column_stack): The principal wants this data side-by-side. Use np.column_stack to merge the IDs, the Scores, and the Statuses into one single 2D array called final_report. Print it to the screen.

4. Data Types: The administration suddenly wants you to add 5 bonus points to the scores in your final_report. Try to add 5 to the score column. What happens? Print the .dtype (data type) of your final_report to figure out why.

(OPTIONAL)

5. The Export: Even though the math broke, save your final_report to your computer as a CSV file called "edupi_final_grades.csv" so another program can deal with it.

```python
# 1. The Numerical Data: Access the 1D marks arrays of 5 Student IDs (101 to 105), store separately in variables . 

student_ids = np.arange(101, 106)
print(student_ids)

scores = np.array([78, 45, 62, 30, 90])
print(scores)
```

```python
# 2. The Text Data: Use `np.where` to create a status array. 
# If the score is 50 or above, the status is "Pass". If it's below 50, it is "Fail".

status = np.where(scores >= 50, "Pass", "Fail")

print("Status:\n", status)
```

```python
# 3. The Merge (column_stack): The principal wants this data side-by-side. 
# Use np.column_stack to merge the IDs, the Scores, and the Statuses into one single 2D array called final_report. Print it to the screen.

final_report = np.column_stack((student_ids, scores, status))

print("Final Report:\n", final_report)
print("\nData type (dtype):", final_report.dtype)
```

```python
# 4. Data Types: The administration suddenly wants you to add 5 bonus points to the scores in your final_report. 
# Try to add 5 to the score column. What happens? Print the .dtype (data type) of your final_report to figure out why.

scores_plus_5 = scores + 5

print("\nScores after +5:", scores_plus_5)

# Try modifying final_report (this is where the issue appears)
try:
    final_report[:, 1] = final_report[:, 1].astype(int) + 5
except Exception as e:
    print("\nError when modifying final_report:", e)

print("\nFinal Report after attempt:\n", final_report)
print("\nFinal dtype:", final_report.dtype)
```

```python
# 5. The Export: Even though the math broke, 
# save your final_report to your computer as a CSV file called "edupi_final_grades.csv" so another program can deal with it.

np.savetxt(
    "edupi_final_grades.csv",
    final_report,
    delimiter=",",
    fmt="%s"
)

print("\nCSV file saved: edupi_final_grades.csv")
```
