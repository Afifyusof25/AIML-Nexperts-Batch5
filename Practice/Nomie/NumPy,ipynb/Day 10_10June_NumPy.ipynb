- In programming, an environment is the setup or context in which your code runs.
- It contains the tools, libraries, settings, and dependencies needed for a program to work.
- It is an isolated workspace where we can install and manage packages without affecting the global Python installation.
- when building software or projects, which version of python we using is important
  for eg.
  if we write some LOC in python 3.10
  the same line may not work in python 3.13
  the developers could hv removed something (deprecated) that LOC

```python
num_list = [1,2,3,4,5,6,7,8,9]

num_list10 = []

for num in num_list:
    num_list10.append(num*10)

print(num_list10)

```
