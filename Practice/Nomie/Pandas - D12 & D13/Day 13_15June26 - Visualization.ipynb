```python
### exercise

# - build a scatterplot of 'total bill' vs 'tip' in the tips dataset

df_tips = sns.load_dataset('tips')

# create scatterplot
sns.scatterplot(data=df_tips,
                x='total_bill',
                y='tip',
                s = 50,
                hue = 'day')

plt.title('Scatterplot of Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```python
# Histogram with Seaborn
sns.histplot(data = df_tips['tip'], bins = 20, edgecolor = "#E20000", kde = True, color = 'blue')
plt.title('Normal Distribution')   # title
plt.xlabel('data')   # label x axis
plt.show()
```

```python
## Subplots:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataset for comparison
data = pd.DataFrame({
    'Age': [23, 45, 34, 42, 25, 36, 52, 39, 40, 30],
    'Salary': [40000, 80000, 60000, 75000, 42000, 62000, 95000, 71000, 85000, 53000]
})

# Scatter Plot of Age vs Salary
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.show()

# Histogram of Age
sns.histplot(data['Age'], bins=5, kde=True, color='green')
plt.title('Histogram of Age')
plt.show()

# Box Plot of Salary
sns.boxplot(data['Salary'], color='orange')
plt.title('Box Plot of Salary')
plt.show()
```

```python
# Create a figure with 1 row and 3 columns
plt.figure(figsize=(15, 5)) # This creates a figure with a specific size, here 15 inches wide and 5 inches tall, to accommodate three plots side by side.

# Scatter Plot (1st subplot)
plt.subplot(1, 3, 1)  # (rows, columns, position)
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.title('Age vs Salary (Scatter Plot)')
plt.xlabel('Age')
plt.ylabel('Salary')

# Histogram (2nd subplot)
plt.subplot(1, 3, 2)  # (rows, columns, position)
sns.histplot(data['Age'], bins=5, kde=True, color='green')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Box Plot (3rd subplot)
plt.subplot(1, 3, 3)  # (rows, columns, position)
sns.boxplot(data['Salary'], color='orange')
plt.title('Box Plot of Salary')
plt.xlabel('Salary')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
plt.figure(figsize=(12, 4))

# Scatter Plot (1st subplot)
plt.subplot(1, 3, 1)  # (rows, columns, position)
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')

# Histogram (2nd subplot)
plt.subplot(1, 3, 2)  # (rows, columns, position)
sns.histplot(data['Age'], bins=5, kde=True, color='green')

# Box Plot (3rd subplot)
plt.subplot(1, 3, 3)  # (rows, columns, position)
sns.boxplot(data['Salary'], color='orange')
```
