```python
##  noms - p1

# Tkinter --- Built-in python - no new installation is needed
import tkinter as tk # aliases 

# create a blank window
window = tk.Tk() # class is used to create the main application window
window.title("My First GUI") # title of the window
window.geometry("400x300") # size of the window (width x height)

# created some text, to attach with the window
greeting = tk.Label(window, 
                    text = "Hello User Interface!",
                    font = ("Arial", 16, "bold"), # font and size
                    fg = "white",
                    bg = "blue"
                     )             

# packing - take the greeting, and put it on the window
greeting.pack()

# action
def button_fool():
    print("You clicked Fool Me, you FOOL!")


# Button
action_button = tk.Button(window,
                          text= "Fool Me",
                          command= button_fool,
                          font=("Helvetica", 16, "bold"), # font, fontsize, style
                          fg= "white", # foreground
                          bg = "red", # background,
                          width= 15,
                          height= 3,
                          relief= "raised", # flat, raised, sunken, solid, ridge, groove
                          cursor= "hand2"   # change the mouse to a hand
                          )
action_button.pack(pady = 15)


# action
def button_click():
    print("You clicked ME!")

# Button
action_button = tk.Button(window,
                          text= "Click Me",
                          command= button_click,
                          font=("Helvetica", 16, "bold"), # font, fontsize, style
                          fg= "black", # foreground
                          bg = "yellow", # background,
                          width= 15,
                          height= 3,
                          relief= "raised", # flat, raised, sunken, solid, ridge, groove
                          cursor= "hand2"   # change the mouse to a hand
                          )
action_button.pack(pady = 15)


# action
def button_praise():
    print("All Hail the Button Clickers!")


# Button
action_button = tk.Button(window,
                          text= "Praise Me",
                          command= button_praise,
                          font=("Helvetica", 16, "bold"), # font, fontsize, style
                          fg= "white", # foreground
                          bg = "blue", # background,
                          width= 15,
                          height= 3,
                          relief= "raised", # flat, raised, sunken, solid, ridge, groove
                          cursor= "hand2"   # change the mouse to a hand
                          )
action_button.pack(pady= 15)



# Keeps the window open
window.mainloop()
```
