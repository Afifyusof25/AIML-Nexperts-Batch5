```python
import tkinter as tk

window = tk.Tk() # create a window
window.title("Calculator by Idrees") # window title
window.geometry("500x500") # window size - width x height

# First input
label1 = tk.Label(window, text = "Enter first Number", font=("Helvetica", 12) )
label1.pack()

input1 = tk.Entry(window)
input1.pack(pady=10)

# Second input
label2 = tk.Label(window, text = "Enter second Number", font=("Helvetica", 12) )
label2.pack()

input2 = tk.Entry(window)
input2.pack(pady = 10)


# fucntion
def calculate():
    try:
        num1 = float(input1.get())
        num2 = float(input2.get())

        add = num1 + num2
        sub = num1 - num2
        mul = num1 * num2
        div = num1 / num2

        output_label.config(text= f"Addition: {add}\nSubtraction: {sub}")

    except Exception as e:
        output_label.config(text= f"Whoa, there is an error: {e}", justify='left')

    
# Calculate
calculate_button = tk.Button(window, text= "Calculate",
                          command= calculate,
                          font=("Helvetica", 16, "bold"), # font, fontsize, style
                          width= 8,
                          height= 2,
                          relief= "raised", # flat, raised, sunken, solid, ridge, groove
                          cursor= "hand2"   # change the mouse to a hand 
                          )
calculate_button.pack(pady=20)

# Output Label
output_label = tk.Label(window, text = "RESULTS", font=("Helvetica", 12),fg = "green", justify= 'left' )
output_label.pack()


window.mainloop()
```

```python
import tkinter as tk

# Function to display button values
def click(value):
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(0, current + str(value))

# Function to calculate the result
def calculate():
    try:
        result = eval(entry.get())
        entry.delete(0, tk.END)
        entry.insert(0, result)
    except:
        entry.delete(0, tk.END)
        entry.insert(0, "Error")

# Function to clear the screen
def clear():
    entry.delete(0, tk.END)


# Main window
window = tk.Tk()
window.title("Simple Calculator")
window.geometry("300x400")

# Display box
entry = tk.Entry(window, font=("Arial", 20), justify="right")
entry.pack(fill="both", ipadx=8, ipady=15, padx=10, pady=10)

# Buttons
buttons = [
    ['7', '8', '9', '/'],
    ['4', '5', '6', '*'],
    ['1', '2', '3', '-'],
    ['C', '0', '=', '+']
]

# Create buttons
for row in buttons:
    frame = tk.Frame(window)
    frame.pack(expand=True, fill="both")

    for button in row:
        if button == "=":
            cmd = calculate
        elif button == "C":
            cmd = clear
        else:
            cmd = lambda b=button: click(b)

        tk.Button(
            frame,
            text=button,
            font=("Arial", 18),
            command=cmd
        ).pack(side="left", expand=True, fill="both")


window.mainloop()
```
