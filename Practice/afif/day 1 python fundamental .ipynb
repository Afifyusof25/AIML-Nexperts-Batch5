# Variables in Python

- variable = labeled box for a value
- you can store any data type in this box
- Variable is a reserved memory location
  - python automatically determines the type of the thing you want to save, and assigns memory
  -


- Rules:
  1. use proper names not abbrevations - for Readable
  2. use snake_case
  3. use camelCase for Functions
  4. use lowercase letters and underscores
  5. avoid abbrevations like - sc, r, cn
  6. A variable name cannot start with a number

```python
# use a method called input to declare variables too
# by taking input

# BMI CALCULATOR

name = input("enter your Name:")
age = int(input("enter your age:"))
weight = int(input("enter your weight (in kg):"))
height = int(input("enter your height (in cm):")) * 0.01

bmi = weight / (height * height)

print(f"Hi {name}! Your BMI is {bmi:.2f}")
```
