## SET  { }
- 
- is an unordered collection of unique items
- very useful when you want to store items that should not repeat

```python
set_A = {1,2,3,4,4,3,2}
print(set_A)


set_B = {3,4,8,5,9,6}
print(set_B)
```

```python
# Set Operations
set_A | set_B # Union - is combination of all elements uniquely

set_A & set_B # Intersection - the common elements in the sets
```

```python
# list

names = ["idrees", "afif", "hazira", "idrees", "Luqman", "hazira"]
print(names)

# remove all the dupicates 
names = set(names)
names = list(names)
print(names)


list(set(names))
```

```python
set_A.add(42) # add an element

set_A.update(names) # addd any iterables

set_A.remove(42) # remove an element

set_A.discard(42) # when not sure if element is already removed
# this does not throw an error


random_element = set_A.pop() # randomly returns an element

set_A.clear()
```

### Task 3: Create a dictionary of 3 countries and their capitals, then:

    - Add a new country and its capital to the dictionary.
    - Remove one country.
    - Update the capital of an existing country.
    - Print the dictionary.

```python
# 1. Create a dictionary of 3 countries and their capitals
countries = {
    "Malaysia": "Kuala Lumpur",
    "Singapore": "Singapore",
    "Thailand": "Bangkok"
}

# 2. Add a new country (Vietnam)
countries["Vietnam"] = "Hanoi"

# 3. Remove one country ( Singapore)
countries.pop("Singapore")

# 4. Update the capital of an existing country ( Malaysia)
countries["Malaysia"] = "Putrajaya"

# 5. Print the dictionary
print(countries)
```
