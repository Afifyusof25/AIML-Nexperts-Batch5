### Exercise - 

- i want to store students in a database
- class for students, and then 4 objects for 4 students
- class attribute: total Studetns
- object attributes: name, age, fav_sub
- class methods: 
    - method for student count - which just says hello and then tells how many students there are
    - 

````python
class Student:
    # Class attribute
    total_students = 0

    # Constructor
    def __init__(self, name, age, fav_sub):
        self.name = name
        self.age = age
        self.fav_sub = fav_sub

        # Increase student count whenever a new object is created
        Student.total_students += 1

    # Class method
    @classmethod
    def student_count(cls):
        print("Hello!")
        print(f"Total number of students: {cls.total_students}")


# Create 4 student objects
student1 = Student("Afif", 22, "Python")
student2 = Student("Alice", 21, "Mathematics")
student3 = Student("Bob", 23, "Physics")
student4 = Student("Charlie", 20, "Machine Learning")

# Print student information
print(student1.name, student1.age, student1.fav_sub)
print(student2.name, student2.age, student2.fav_sub)
print(student3.name, student3.age, student3.fav_sub)
print(student4.name, student4.age, student4.fav_sub)

# Call the class method
Student.student_count()
````

### Inheritance and Polymorphism
Inheritance
- a new class (CHILD) can reuse code from a existing class (PARENT)
- 
Polymorphism
- the same method, gives different outputs

##### Inheritance
```python
super().__init__()  # is used to call a parent class
```
- example:
- Parent===== Animal
- Children Classes === Dog, Cat etc

````python
# Parent class
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self._age = age 

    def introduce(self):
        print(f"Hi. I am {self.name}!!\nI am a {self.species}\nI am {self._age} years old!")

    # Getter Method
    def get_age(self):
        return self._age
    
    # Setter Method
    def set_age(self, new_age):
        self._age = new_age
        return f"The new age is {new_age}"
    
    def have_birthday(self):
        self._age +=1
        print(f"It's {self.name}'s birthday. Now {self._age} years old ")
````

````python
# CHILD of ANIMAL
class Dog(Animal):

    def __init__(self, name, age): # this is what we will pass ourselves
        # super() refers to the parent class --- could have used Animal as well
        # 
        self._age = age
        super().__init__(name, "Dog", age)

    def sound(self):
        print(f"{self.name} says: Woof!!")

    # different methods for different classes
    def detect_drugs(self):
        if self._age > 5:
            return f"I will sniff out any drugs"
        else:
            return f"{self.name} says, I am just {self.age} years old. You fr??"
````

````python
# CHILD of ANIMAL
class Cat(Animal):

    def __init__(self, name, age):
        super().__init__(name,"Cat", age)

    def sound(self):
        print(f"{self.name} says: Meowww!!")
````

````python
# CHILD of ANIMAL
class Parrot(Animal):

    def __init__(self, name, age):
        super().__init__(name,"Parrot", age)

    def sound(self):
        print(f"{self.name} says: Mithu Kakaaa!!")
````

Exercise 1: Coffee Machine

Create a CoffeeMachine class.

In the `__init__`, set up two "protected" attributes: `_water_level` (starts at 100) and _bean_level (starts at 50).

Create a make_coffee() method. It should check if there is enough water (needs 20) and beans (needs 10).

If yes, deduct the amounts and print "Enjoy your coffee!"

If no, print "Please refill resources."

Create refill_water() and refill_beans() methods to top the machine back up.

````python
class CoffeeMachine:
    def __init__(self):
        # Protected attributes
        self._water_level = 100
        self._bean_level = 50

    def make_coffee(self):
        if self._water_level >= 20 and self._bean_level >= 10:
            self._water_level -= 20
            self._bean_level -= 10
            print("Enjoy your coffee!")
        else:
            print("Please refill resources.")

    def refill_water(self):
        self._water_level = 100
        print("Water tank refilled.")

    def refill_beans(self):
        self._bean_level = 50
        print("Bean container refilled.")

    def show_levels(self):
        print(f"Water Level: {self._water_level}")
        print(f"Bean Level: {self._bean_level}")


# Create object
machine = CoffeeMachine()

# Check levels
machine.show_levels()

# Make coffee
machine.make_coffee()
machine.make_coffee()

# Check levels again
machine.show_levels()

# Refill
machine.refill_water()
machine.refill_beans()

# Check levels
machine.show_levels()
````

Exercise 2: The RPG Hero Roster
Target Concept: Inheritance & Polymorphism (Overriding behaviors)

The Scenario: You are building the combat system for a video game. Every character attacks, but how they attack depends on their class.

Requirements:

Create a base class called Hero with an __init__ that takes a name and health.

Give Hero a method called attack() that simply prints: "[Name] performs a basic attack!"

Create two subclasses: Warrior and Mage that inherit from Hero.

Polymorphism: Override the attack() method in both subclasses.

The Warrior should print: "[Name] swings a heavy broadsword!"

The Mage should print: "[Name] casts a fireball!"

````python
class Hero:
    def __init__(self, name, health):
        self.name = name
        self.health = health

    def attack(self):
        print(f"{self.name} attacks!")


class Warrior(Hero):
    def attack(self):
        print(f"{self.name} swings a sword!")


class Mage(Hero):
    def attack(self):
        print(f"{self.name} casts a fireball!")


# Create objects
hero1 = Hero("Alex", 100)
warrior1 = Warrior("Thor", 150)
mage1 = Mage("Merlin", 80)

# Call methods
hero1.attack()
warrior1.attack()
mage1.attack()
````

Exercise 3: The E-Commerce Cart
Target Concept: Object Interaction (Passing objects into other objects)

The Scenario: A shopping cart isn't just a list of text strings; it's a collection of actual Product objects.

Requirements:

Create a Product class that takes a name and a price.

Create a ShoppingCart class. Its __init__ should just create an empty list: self.items = [].

Give the ShoppingCart an add_item(product) method. Notice that the argument is expected to be a Product object, not just a string!

Add a calculate_total() method to the cart that loops through self.items and adds up all their prices.

````python
# Product class
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price


# Shopping Cart class
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, product):
        self.items.append(product)
        print(f"{product.name} added to cart.")

    def calculate_total(self):
        total = 0

        for product in self.items:
            total += product.price

        print(f"Total Price: RM{total}")


# Create products
product1 = Product("comic", 10.00)
product2 = Product("pencil", 8.50)
product3 = Product("Notebook", 15.00)

# Create shopping cart
cart = ShoppingCart()

# Add products to cart
cart.add_item(product1)
cart.add_item(product2)
cart.add_item(product3)

# Calculate total
cart.calculate_total()
````
