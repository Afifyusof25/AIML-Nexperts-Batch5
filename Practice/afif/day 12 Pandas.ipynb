Adding and Modifying columns

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Department': ['Math', 'Chemistry', 'Biology', 'Engineering', 'History'],
    'Salary': [50000, 60000, 55000, 62000, 58000],
    'Experience': [3, 5, 2, 4, 6]
})

df['Email'] = [
    'Alice@gmail.com',
    'Bob@gmail.com',
    'Charlie@gmail.com',
    'Diana@gmail.com',
    'Eve@gmail.com'
]

df.loc[2, 'Department'] = 'Physics'
df.Salary = df.Salary / 1000
df.Experience = df.Experience * 2

print(df)

df
```
