3. Vowel Counter: You asked a user to input a word. Count the number of vowels (a,e,i,o,u) in the word

```python
word = input("Type your word or sentence").lower()

# count the number of vowels
vowel_count = word.count('a') + word.count('e') +word.count('i') +word.count('o') +word.count('u')
print("Number of vowels: ", vowel_count)
```

```python
word = input("Type you word here")
 
vowels = "aeiou"
count = 0
 
for i in word.lower():
    if i in vowels:
        count += 1 # increment
 
print("Number of vowels:", count)
```

5. FizzBizz program: Ask user for input.
    - Numbers range from 0 to 20
    - if input is divisible by 3, print 'Fizz'
    - if divisible by 5, print 'Buzz'
    - if divisible by 3 and 5 both, print 'FizzBuzz'

```python
number = int(input("Enter a number between 0 to 20: "))
word = ""

if (number % 3) == 0 and (number % 5) == 0:
  word = "FizzBuzz"
elif (number % 5) == 0:
  word = "Buzz"
elif (number % 3) == 0:
  word = "Fizz"
else:
  word = "Invalid"
 
print(word)
```
