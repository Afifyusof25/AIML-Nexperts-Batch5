```python
# 4 - add 5 bonus points to the scores in your final_report.

import numpy as np

student_ids = np.array([101, 102, 103, 104, 105])
scores = np.array([75, 42, 88, 50, 30])

status = np.where(scores >= 50, "Pass", "Fail")

final_report = np.column_stack((student_ids, scores, status))

print(final_report)

print("\nData Type:")
print(final_report.dtype)

score_column = final_report[:, 1].astype(int)

print("\nBonus +5:")
print(score_column + 5)
```
