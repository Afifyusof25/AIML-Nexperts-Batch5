## Keyword Arguments and Default Arguments
- *kwargs
- you give a name to the argument

````python
def summing(num1, num2, power): # parameters
    sum = (num1 + num2) ** power
    return sum


# function call
summing(1, -2, 3) # order is important
# simple arguments

# keyword arguments --- Order is immaterial/not important
summing(power=3, num1= -3, num2= 1) # but there is a keyword to them as well
summing(1,2,3)

# ------------------------------------------
# with default arguments
def summing(num1 = 1, num2 = 1, power = 1): # parameters
    sum = (num1 + num2) ** power
    return sum


summing(3,2,2)
````

````python
# Create a function that gives the sum of n numbers
# Sum of first n numbers ----> (n* (n +1)) // 2
# sum_n(10) -> 1 + 2 + 3 + ...+ 10 = 

def sum_n(n):
  sum = 0
  for i in range(n+1):
    sum = sum + i
  return sum

sum_n(100_000)
````

````python
# function that takes an number, and returns "Even" or "Odd"

def even_odd(num):

    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
    
even_odd(53)
````

````python
# Prime number function
# We pass a number, and the function says if its prime or not
 
def prime(num):
    if num == 1 or num == 2 or num == 3:
        print('Prime Number')
    elif num % 2 == 0 or num % 3 == 0:
        print('Non Prime Number')
    elif num % 1 == 0 and num % num == 0:
        print ('Prime Number')

prime(23)

# check all the prime numbers upto 30
# 2 Prime
# 3 Prime
# 4 Non Prime
````

2. Repeat until user enters 'q' or 'quit'

````python
def calculator():
    while True:
        user_input = input("Enter two numbers separated by space (or q to quit): ")

        if user_input.lower() in ["q", "quit"]:
            print("Goodbye!")
            break

        a, b = map(float, user_input.split())

        print("Sum:", a + b)
        print("Difference:", a - b)
        print("Multiplication:", a * b)
        print("Division:", a / b)

calculator()
````

## Function Recursion

- function calls itself
- Remember:
    - Base Case: when to stop
    - Recursive case: do a small part, and call function

```python
def function(<parameters>):
    if <condition>:
        return result
    else: 
        return function(<modified parameters>)

````python
https://www.w3schools.com/python/python_recursion.asp
````

````python
# 10 -> 0
def countdown(n):
    print(n)
    # base case - to exit the infinite loop
    if n == 0:
        print("DONE!!!")
        return n

    # recursive case - with updated parameters
    countdown(n - 1) # parameters are updated
````

````python
countdown(10)
````

````python
n = 5
factorial = 1
for i in range(1, n+1):
    factorial = factorial * i

print(factorial)
````

````python
# factorial using for loop
def multiplyn(n):
    factorial = 1
    for i in range(1, n+1):
        factorial = factorial * i
    return factorial

multiplyn(10)
````

````python
multiplyn(20)
````

````python
# factorial using for loop
def multiplyn(n: int):
    # base case
    if n == 1:
        return n
    
    # recursive
    return n * multiplyn(n - 1)

multiplyn(5.0)
````

### create a timer function using recursion
- start from 10 go to 0 - or vice versa

````python
import time

def timer(count):
    print(count)

    if count == 0:
        print("Time's up!")
        return

    time.sleep(1)
    timer(count - 1)

timer(10)
````
